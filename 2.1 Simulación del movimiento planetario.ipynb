### 2.1 Simulación del movimiento planetario

El algoritmo de Verlet es una excelente opción para programar simulaciones dada su simplicidad y precisión, por lo tanto se utilizará en algunas aplicaciones.

El empleo de este algoritmo se ha enfocado principalmente a sistemas microscópicos, de modo que el cálculo de trayectorias de móviles en presencia de fuerzas gravitacionales, dadas las dimensiones involucradas es una aplicación de interés. En este caso se desarrollara la trayectoria de la Tierra alrededor del Sol

Usaremos unidades astronómicas para la longitud (1 ua es el radio de la órbita terrestre), años para el tiempo y masas solares (mS) como unidad de masa. Ahora bien, para el cálculo de la constante gravitacional $G$ en las unidades elegidas, se empleará la Tercera Ley de Keppler, la cual relaciona al periodo de un planeta con el radio de su órbita y con la masa del planeta atrayente:

\begin{align*}
T^2 = \frac{4\pi^2a^3}{GM_0} \tag{2.1}
\end{align*}

En este caso tomamos el radio de la órbita terrestre (1 ua) y el periódo de su órbita al rededor del sol (1 año), y la masa del sol (1 mS), con esto despejamos $G$ obteniendo:

\begin{align*}
G = 4\pi^2\left[\frac{ua^3}{año^2 * mS}\right] \tag{2.2}
\end{align*}

Ahora debemos calcular las condiciones iniciales (velocidad y posición) para la simuación en base a los datos que aparecen en los textos comunes, esto es periodo ($T$), excentricidad ($\epsilon$) y la masa. Para facilitar el, manejo algebráico definamos la siguente cantidad como:

\begin{align*}
k= GMm \tag{2.3}
\end{align*}

Donde $M$ es la masa de los planetas que intervienen en el sistema, de modo que $M$ sería la masa del sol y $m$ la masa del planeta el cual deseamos obtener su trayectoria, de modo que en nuestras unidades $k=4\pi^2m$ 

R.Q. Hockey desarrolla una relación entre las coordenadas polares que describen la posición de una partícula en función de las magnitudes dinámicas del movimiento, la cual está dada como sigue:

\begin{align*}
\left( \frac{dr}{d\theta} \right)^2 = \frac{m^4r^4}{L^4} \left( \frac{2[E-E_p(r)]}{m}-\frac{L^2}{m^2r^2} \right) \tag{2.4}
\end{align*}

Donde $L$ es el movimiento angular de la partícula y $E_p$ es la energía potencial. En coordenadas polares, cuando el foco está en el orígen, la ecuación de la elipse con distancia del foco a la directriz $d$ es:

\begin{align*}
\frac{\epsilon d}{r} = 1+\epsilon\cos(\theta) \tag{2.5}
\end{align*}

O bien:

\begin{align*}
\sin^2(\theta) = 1 - \frac{d^2}{r^2} + \frac{2d}{\epsilon r} - \frac{1}{\epsilon^2} \tag{2.6}
\end{align*}

Derivando la ecuación anterior se obtiene:

\begin{align*}
\left( \frac{dr}{d\theta} \right)^2 = \frac{r^4\sin^2(\theta)}{d^2} \tag{2.7}
\end{align*}

y sustituyendo este resultado en la ecuación (2.4) se llega a lo siguiente:

\begin{align*}
\sin^2(\theta) = \frac{d^2m^2}{L^2} \left( \frac{2[E-E_p(r)]}{m}-\frac{L^2}{m^2r^2} \right) \tag{2.8}
\end{align*}

Al sustuir en la ecuación (2.6) se obtiene:

\begin{align*}
1-\frac{d^2}{r^2}+\frac{2d}{\epsilon r} - \frac{1}{\epsilon^2} = \frac{2d^2mE}{L^2} - \frac{2d^2mE}{L^2} - \frac{2d^2mE_p(r)}{L^2} - \frac{d^2}{r^2} \tag{2.9}
\end{align*}

Y agrupando bien los términos, tenemos:

\begin{align*}
1-\frac{1}{\epsilon^2} - \frac{2d^2mE}{L^2} = -\frac{2d}{\epsilon r} - \frac{2d^2mE_p(r)}{L^2} \tag{2.10}
\end{align*}
 
 Podemos igualar entre sí los términos constantes, así como los dependientes de $r$ para obtener lo siguiente:
 
\begin{align*}
1-\frac{1}{\epsilon ^2} = \frac{2d^2mE}{L^2} \tag{2.11}
\end{align*}

y 

\begin{align*}
-\frac{2d}{\epsilon r} = \frac{2d^2mE_p(r)}{L^2} \tag{2.12}
\end{align*}

de donde 

\begin{align*}
E = \frac{L^2}{2d^2m} \left( 1- \frac{1}{\epsilon^2} \right) = -\frac{L^2}{2d^2m} \left(\frac{1-\epsilon^2}{\epsilon^2} \right) \tag{2.13} 
\end{align*}

y 

\begin{align*}
E_p(r) = -\frac{L^2}{md\epsilon r} \tag{2.14}
\end{align*}

Sabemos que el semieje mayor $a$ de una elipse puede expresarse como:

\begin{align*}
a = \frac{\epsilon d}{1-\epsilon^2} \tag{2.15}
\end{align*}

Despejando $d$ y sustituyend en (2.13):

\begin{align*}
E = - \frac{L^2}{2\epsilon dma} \tag{2.16}
\end{align*}

Ahora bien, sabemos que $E_p(r) = -G \tfrac{mM}{r}, entonces al sustituir este valor en (2.14) obtenemos:

\begin{align*}
\frac{L^2}{m\epsilon d} = GMm \tag{2.17}
\end{align*}

y por último, comparando esta expresión con la ecuación (2.16):

\begin{align*}
E = - G \frac{mM}{2a} = -\frac{k}{2a} \tag{2.18}
\end{align*}

Entonces despejando $a$ de la ecuación (2.1) y sustituyendo en (2.18) tenemos:

\begin{align*}
E = -\frac{k}{2\left( \frac{T^2k}{4\pi^2m} \right)^{\frac{1}{3}}} \tag{2.19}
\end{align*}

Con lo cual, obtenemos lo siguiente:

\begin{align*}
E = - \left(\frac{\pi^2k^2}{2T^2}m\right)^{\frac{1}{3}} \tag{2.20}
\end{align*}

Agora, de (2.15) despejamos $\epsilon d$ y al sustituir en (2.17) obtenemos: 

\begin{align*}
L^2 = -Gm^2Ma(1-\epsilon^2) \tag{2.21}
\end{align*}

y sustituyendo el valor $a$ mediante (2.15), otenemos la expresión para la excentricidad: 

\begin{align*}
\epsilon = \sqrt{1+ \frac{2EL^2}{mk^2}} \tag{2.22}
\end{align*}

de aquí:

\begin{align*}
L = \sqrt{\frac{mk^2(\epsilon^2-1)}{2E}} \tag{2.23}
\end{align*}

Por otro lado, minimizando la expresión de la trayectoria en coordenadas polares, se llega a lo siguiente:

\begin{align*}
r_{\text{min}} = \frac{L^2}{(1+\epsilon) mk} \tag{2.24}
\end{align*}

Con lo cual, introduciendo esta expresión, las cantidades antes calculadas, obtenemos la posición inicial para nuestro problema que corresponde con el punto donde la velocidad horizontal se anula.

Por otro lado, expresando la energía total como la suma de energías cinética y potencial:

\begin{align*}
E = \frac{1}{2}mv_{y}^2 + \frac{k}{r_{\text{min}}} \tag{2.25}
\end{align*}

podemos obtener la velocidad inicial:

\begin{align*}
v_{0}^y = \sqrt{  \frac{2}{m}  \left( E + \frac{k}{ r_{ \text{min} } } \right) } \tag{2.26}
\end{align*}

Ahora bien, la fuerza central que interviene en el sistema está dada por:

\begin{align*}
\vec{F} = G\frac{mM}{r^2}\hat{r} \tag{2.27}
\end{align*}

donde $\vec{r}$ es el vector que une a ambas masas y $\hat{r}$ es un vector unitario en la dirección de $\vec{r}$.

Entonces, dado que el sistema es bidimensional, el esquema del algoritmo de Verlet sería el siguiente:

\begin{align*}
x_{n+1} = 2x_n - x_{n-1} - h^2 \frac{GM}{r_{n}^2} \left( \frac{x_n}{r_n} \right) \\
y_{n+1} = 2y_n - y_{n-1} - h^2 \frac{GM}{r_{n}^2} \left( \frac{y_n}{r_n} \right) \tag{2.28} 
\end{align*}

Con $r_n^2 = x_n^2 + y_n^2$ 

Para la inicialización del algorítmo, se emplea la ecuación (2.39), de modo qué:

\begin{align*}
x_{-1} = x_0 - v_0^xh + \frac{h^2}{2m}F(x_0,y_0) \\
y_{-1} = y_0 - v_0^yh + \frac{h^2}{2m}F(x_0,y_0) \tag{2.29} 
\end{align*}

Con $v_0^x =0$, $y_0 = 0$ y $x_0 = r_{\text{min}}$.

El valor de $h$ estará dado por el tamaño de intervalos en los cuales se dividirá el periodo de la órbota, lo cual sera un factor importante para la precisión del cálculo, esto es, que al hacer divisiones más finas podremos obtener mejores resultados. Se mostrará más adelante que es posible obtener una predicción para el tiempo en que la Tierra está en una posición dada con una posición dada con una precisión de 30 segundos. Sin embargo, nuevamente aparece el cpmpromiso de tener una simulación más ágil (en términos de velocidad de cálculo y visualización) o una simulación precisa, dado que la velocidad de procesamiento del equipo de cómputo será soempre una limitación.

A continuación se muestra el código que presenta una animación de movimiento planetario y muestra una tabulación de posición contra tiempo.

```python
import numpy as np
import matplotlib.pyplot as plt

#Parámetros
h = 1.0 / 365.0 #El intervalo de tiempo es un día
m = 1.0 / 332946.0 #La masa de la Tierra
ep = 0.017 #Excentricidad
T = 1.0 #Periodo
#Constantes calculadas en términos de los parámetros
G = 4.0*np.pi**2 
k = G*m 
E = -np.power( ( (np.pi**2)*(k**2)*m )/(2*T**2), 1/3 )
L2 = (m*(k**2)*(ep**2-1))/(2*E)
rmin = L2/( (1+ep)*m*k )

#Condiciones iniciales para x, y
x1 = rmin
y1 = 0.0
#Condiciones iniciales para vx, vy
vx = 0.0
vy = np.sqrt(2/m*(E+k/rmin))

# Funciones de fuerza
def fx(x, y):
    r3 = np.power(x**2 + y**2, 1.5)
    return -G/r3*x
def fy(x, y):
    r3 = np.power(x**2 + y**2, 1.5)
    return -G/r3*y

#Declaración de los vectores donde se guardarán los valores
X = np.zeros(365)
Y = np.zeros(365)
T = np.zeros(365)

#Calculando los valores x0, y0 a partir de las condiciones iniciales
x0 = x1 - vx*h + h**2*fx(x1, y1)/2
y0 = y1 - vy*h + h**2*fy(x1, y1)/2

#Simulación usando el algoritmo de Verlet
for t in range(365):
    x2 = 2*x1 - x0 + h**2*fx(x1, y1)
    y2 = 2*y1 - y0 + h**2*fy(x1, y1)
    #Almacenando los valores de las variables a medida que se itera t
    X[t] = x2
    Y[t] = y2
    T[t] = t+h
    # Actualizar posiciones
    x0 = x1
    x1 = x2
    y0 = y1
    y1 = y2

#Graficando la trayectoria
plt.figure(figsize=(8,8))
plt.plot(X,Y, label='Órbita de la Tierra')
plt.scatter(0,0, color='orange',s=300, label="Sol")  # Representar al Sol
plt.xlabel('x [UA]')
plt.ylabel('y [UA]')
plt.title("Simulación del movimiento Planetario")
plt.legend()
plt.grid(True)
plt.show()
```

Veamos ahora un planteamiento analítico del problema, para evaluar el desempeño del método numérico.

Partiendo de la condición de conservación del momento angular:

\begin{align*}
\frac{d\theta}{dt} &= \frac{L}{mr^2}\\
\Rightarrow \int_{\theta_0}^{\theta} mr^2 d\theta &= \int_{t_0}^{t} L \text{ } dt \tag{2.30}
\end{align*}

En el cual es posible plantear $r$ en función de $\theta$ mediante la ecuación de una sección cónica en coordenadas polares:

\begin{align*}
r = \frac{\epsilon d}{1+\epsilon \cos (\theta)} \tag{2.31}
\end{align*}

donde $d$ es una constante dada por la distancia del foco a la directriz y puede obtenerse mediante la siguiente relación:

\begin{align*}
d = r_{\text{min}} \left( 1+\frac{1}{\epsilon} \right) \tag{2.32}
\end{align*}

Sustituyendo la ecuación de la cónica en la ecuación (2.30) se tiene:

\begin{align*}
m\epsilon^2 d^2 \int_{\theta_0}^{\theta} \frac{d\theta}{[1+\epsilon \cos(\theta)]^2} - L (t-t_0) = 0
\end{align*}

Considerando $\theta_0=0$ y $t_0=0$ entonces:

\begin{align*}
m\epsilon^2 d^2 \int_{0}^{\theta} \frac{d\theta}{[1+\epsilon \cos(\theta)]^2} - Lt = 0 \tag{2.33}
\end{align*}

y 

\begin{align*}
m\epsilon^2 d^2 \int_{0}^{\theta} \frac{d\theta}{[1+\epsilon \cos(\theta)]^2} = 2\text{arctan}(\theta) \left( \frac{(\epsilon-1)\tan(\tfrac{\theta}{2})}{\sqrt{1-\epsilon^2}} \right) \left( \frac{m\epsilon^2 d^2}{\sqrt{1-\epsilon^2}(\epsilon^2-1)} \right)+\frac{m\epsilon^2d^2\sin\theta}{(\epsilon^2-1)(1+\epsilon \cos\theta)} \tag{2.34}
\end{align*}

Entonces con la ecuación (2.33) se tiene que:

\begin{align*}
2\text{arctan}\left( \frac{(\epsilon-1)\tan(\tfrac{\theta}{2})}{\sqrt{1-\epsilon^2}} \right) \frac{m\epsilon^2d^2}{\sqrt{1-\epsilon^2}(\epsilon^2-1)} + \frac{m\epsilon^2d^2\sin(\theta)}{(\epsilon^2-1)(1+\epsilon\cos\theta)} - Lt = 0 
\end{align*}

Por lo cual: 

\begin{align*}
t=\frac{1}{L}\left[2\text{arctan}\left( \frac{(\epsilon-1)\tan(\tfrac{\theta}{2})}{\sqrt{1-\epsilon^2}} \right) \frac{m\epsilon^2d^2}{\sqrt{1-\epsilon^2}(\epsilon^2-1)} + \frac{m\epsilon^2d^2\sin(\theta)}{(\epsilon^2-1)(1+\epsilon\cos\theta)}\right] \tag{2.35}
\end{align*}

Ahora empleamos al algoritmo de verlet para predecir el tiempo en el que ocurre algún evento, como puede ser alguna posición específica, para ello, es necesario hacer un cálculo analítico directo con los datos correspondientes a la Tierra para comparar posteriormente con el resultado que arroje el procedimiento numérico. Por ejemplo mediante la ecuación (2.35) podemos facilmente obtener $t=\tfrac{\pi}{4}$, lo cual corresponde al equinoccio de Otoño y se reduce a :

\begin{align*}
t = \frac{1}{L} \left[ 2\text{arctan} \left( \frac{\epsilon -1}{\sqrt{1-\epsilon^2}} \right) \right]\frac{m\epsilon^2d^2(1+\epsilon\sqrt{1-\epsilon^2})}{\sqrt{1-\epsilon^2}(\epsilon^2-1)} \tag{2.36}
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
h = 1.0 / 365.0  # El intervalo de tiempo es un día
m = 1.0 / 332946.0  # La masa de la Tierra
ep = 0.017  # Excentricidad
T = 1.0  # Periodo
# Constantes calculadas en términos de los parámetros
G = 4.0 * np.pi ** 2 
k = G * m 
E = -np.power(((np.pi ** 2) * (k ** 2) * m) / (2 * T ** 2), 1 / 3)
L2 = (m * (k ** 2) * (ep ** 2 - 1)) / (2 * E)
rmin = L2 / ((1 + ep) * m * k)

# Condiciones iniciales para x, y
x1 = rmin
y1 = 0.0
# Condiciones iniciales para vx, vy
vx = 0.0
vy = np.sqrt(2 / m * (E + k / rmin))

# Funciones de fuerza
def fx(x, y):
    r3 = np.power(x ** 2 + y ** 2, 1.5)
    return -G / r3 * x

def fy(x, y):
    r3 = np.power(x ** 2 + y ** 2, 1.5)
    return -G / r3 * y

# Declaración de los vectores donde se guardarán los valores
X = np.zeros(365)
Y = np.zeros(365)
T = np.zeros(365)

# Calcular x0, y0 a partir de las condiciones iniciales
x0 = x1 - vx * h + h ** 2 * fx(x1, y1) / 2
y0 = y1 - vy * h + h ** 2 * fy(x1, y1) / 2

# Ángulo objetivo (en radianes)
theta_target = np.deg2rad(45)  # Por ejemplo, 30 grados en radianes

# Simulación usando el algoritmo de Verlet
for t in range(365):
    x2 = 2 * x1 - x0 + h ** 2 * fx(x1, y1)
    y2 = 2 * y1 - y0 + h ** 2 * fy(x1, y1)
    # Almacenar los valores de las variables a medida que se itera t
    X[t] = x2
    Y[t] = y2
    T[t] = t * h
    # Convertir a coordenadas polares
    r = np.sqrt(x2 ** 2 + y2 ** 2)
    theta = np.arctan2(y2, x2)  # Ángulo en radianes
    # Verificar si el ángulo está cerca del objetivo
    if np.abs(theta - theta_target) < 0.01:  # Tolerancia en radianes
        time_to_theta = T[t]
        print(f"La Tierra alcanza θ = {np.rad2deg(theta_target):.1f}° en t = {time_to_theta:.2f} años")
        break
    # Actualizar posiciones
    x0 = x1
    x1 = x2
    y0 = y1
    y1 = y2

# Graficar la trayectoria
plt.figure(figsize=(8, 8))
plt.plot(X, Y, label='Órbita de la Tierra')
plt.scatter(0, 0, color='orange', s=300, label="Sol")  # Representa al Sol
plt.scatter(x2, y2, color='blue', s=50, label="Sol")  # Representa a la Tierra
plt.xlabel('x [UA]')
plt.ylabel('y [UA]')
plt.title("Simulación del Movimiento Planetario")
plt.legend()
plt.grid(True)
plt.show()
```
